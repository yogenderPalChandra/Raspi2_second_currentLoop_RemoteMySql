```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

X = np.linspace(0,2,1000)
Y = X**2 + np.random.random(X.shape)

fig = plt.figure()
ax = fig.add_subplot(111)
plt.ion()

fig.show()
fig.canvas.draw()

while True:
    Y = X**2 + np.random.random(X.shape)
    df = pd.DataFrame([X, Y]).transpose()
    df.columns=["A", "B"]
    ax.clear()
    ax.plot(df)
    fig.canvas.draw()
    
```
